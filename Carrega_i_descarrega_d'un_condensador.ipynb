**Carrega i descarrega d'un condensador**

En primer lloc carreguem les llibreries que necessitem



```python
import numpy as np
import matplotlib.pyplot as plt
```

Definim la nostra funció

```python
def capacitor_charge_discharge(t,R,C,V0):
   V_charge = V0 * (1 - np.exp(-t / (R * C))) #Carrega del condensador
   V_discharge = V0 * np.exp(-t / (R * C))  # Descarrega del condensador
   return V_charge, V_discharge
```

Definim els paràmetres del nostre circuit RC

```python
R= 1000 #Resistència en ohms (Ω)
C = 0.001  # Capacitància en faradis (F)
V0 = 5  # Voltatge inicial en volts (V)
print(R,C,V0)
```

Definim una llista amb el paràmetre t a estudiar, de 0 a 5 segons, amb intervals de 0.1 segons

```python
t= np.arange(0.0, 5.1, 0.1)
```

Calculem la carrega i la descarrega del condensador

```python
V_carrega, V_descarrega = capacitor_charge_discharge(t, R, C, V0)
```

**Grafiquem els resultats**

```python
plt.plot(t, V_carrega, '#2ecc71', linestyle='-', label="Càrrega del condensador")
plt.plot(t, V_descarrega, '#515a5a', linestyle='-', label="Descàrrega del condensador")
plt.xlabel('temps (s)')
plt.ylabel('Voltatge (V)')
plt.title("Càrrega y descàrrega d'un condensador en un circuit RC")
plt.legend()
plt.show()
```
